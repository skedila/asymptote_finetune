```python
from google.colab import drive
drive.mount('/content/drive')

import os, json
from tqdm import tqdm
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
```

Load Model and Tokenizer

```python
dir_model = "/content/drive/MyDrive/asymptote_model/phi3_lora_trained"
path_test = "/content/drive/MyDrive/asymptote_model/data/data_splits/test.jsonl"
path_out = "/content/drive/MyDrive/asymptote_model/eval/generated_predictions.jsonl"
```

```python
tokenizer = AutoTokenizer.from_pretrained(dir_model)
model = AutoModelForCausalLM.from_pretrained(
    dir_model,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True
)
model.eval()
```

Format Test Data (Phi-3 prompt format)

```python
# Load data
def load_jsonl(path):
    with open(path, "r") as f:
        return [json.loads(line) for line in f]

raw_test = load_jsonl(path_test)

# Format into Phi-3 prompt format
def format_prompt(example):
    return (
        "<|system|>\nYou are a helpful assistant that writes Asymptote code.\n<|end|>\n"
        f"<|user|>\n{example['description']}\n<|end|>\n<|assistant|>\n"
    )
```

Generate Predictions

```python
results = []
for example in tqdm(raw_test, desc="Generating outputs"):
    prompt = format_prompt(example)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        output_ids = model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=False
        )

    generated = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    # Strip prompt from full decoded output
    predicted_code = generated.split("<|assistant|>\n")[-1].split("<|end|>")[0].strip()

    results.append({
        "id": example.get("id", ""),
        "description": example["description"],
        "reference_code": example["asy_code"],
        "generated_code": predicted_code
    })
```

Save Predictions

```python
os.makedirs(os.path.dirname(path_out), exist_ok=True)
with open(path_out, "w") as f:
    for item in results:
        json.dump(item, f)
        f.write("\n")

print(f"Saved {len(results)} predictions to: {path_out}")
```
